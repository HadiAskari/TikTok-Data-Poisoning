```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from collections import Counter, OrderedDict
```

```python
listofcsv=os.listdir("Initialhashtags/a")
```

```python
# for item in listofcsv:
allhashtags=[]
df=pd.read_csv("Initialhashtags/a/fundthepolice.csv")
temp=df['Hashtags'].to_list()
for index,item in enumerate(temp):
        temp1=temp[index]
        temp1=temp1.replace("[", "")
        temp1=temp1.replace("]", "")
        temp1=list(temp1.split(","))
        for k,hashtag in enumerate(temp1):
            allhashtags.append(temp1[k])
        
newallhashtags=[]
for k,v in enumerate(allhashtags):
    newallhashtags.extend(list(allhashtags[k].split("#")))

newallhashtags=[s.replace('fyp','') for s in newallhashtags]
newallhashtags=[s.replace('FYP','') for s in newallhashtags]
newallhashtags=[s.replace('foryou','') for s in newallhashtags]
newallhashtags=[s.replace('foryoupage','') for s in newallhashtags]
newallhashtags=[s.replace('fyp„Ç∑','') for s in newallhashtags]

newnewallhastags=[]
for item in newallhashtags:
    item=item.strip()
    if item == '':
        continue
    if item == " '":
        continue
    if item =="'":
        continue
    if item =="''":
        continue
    newnewallhastags.append(item)

print(newnewallhastags)
```

```python
count=Counter(newnewallhastags)
print(count)
```

```python
finalcount={}
for k,v in count.items():
    if v > 1:
        finalcount[k]=v
```

```python
print(finalcount)
```
